```python
#Creates a dictionary of the format {team1: {player1: [0,0,0], player2: [0,0,0], ...}, team2: {...}, ...}
#The array for each player is [PointsFor, PointsAgainst, Possessions]
teamDict ={team: {} for team in list(GameLineup['Team_id'].unique())}
for team in teamDict.keys():
    players = list(GameLineup[GameLineup['Team_id'] == team]['Person_id'].unique())
    for player in players:
        teamDict[team][player] = np.array([0, 0, 0])
```

```python
#Team Dictionary above might not be useful or convenient. 
#Will always access by players and thus requires another list to keep track of what team a player is on
#Instead of its just a full list of players, it doesn't really matter what their associated team is, 
#as long as game lineups tracked

#Also realized format is rating per game so this should be done a per game basis
playerDict = {player: np.array([0,0,0]) for player in list(GameLineup['Person_id'].unique())}
```

```python
j = 0
def addPoints(player, points, team0Lineup, team1Lineup, gameDict, possession):
    #Player is person1 of the event messages. Every teammate gets the points for added and opponents points against
    global j
    #print(j)
    j+=1
    if player in team0Lineup:
        for team0player in team0Lineup:
            #Add points for and off possession
            gameDict[team0player] += np.array([points, 0, possession, 0])
        for team1player in team1Lineup:
            #Add points against and def posession
            gameDict[team1player] += np.array([0, points, 0, possession])
        return
    elif player in team1Lineup:
        for team0player in team0Lineup:
            #Add points against and def possession
            gameDict[team0player] += np.array([0, points, 0, possession])
        for team1player in team1Lineup:
            #Add points for and off possession
            gameDict[team1player] += np.array([points, 0, possession, 0])
        return
    else:
        print("Player not found on either team lineup!")
        print(player)
        print('Team0', team0Lineup)
        print('Team1', team1Lineup)
        return -1
    
def addPossession(team0Lineup, team1Lineup, gameDict):
    #Assumes team0 is offensive team and team1 on defense
    for team0player in team0Lineup:
        #Add off possession
        gameDict[team0player] += np.array([0, 0, 1, 0])
    for team1player in team1Lineup:
        #Add points against and def posession
        gameDict[team1player] += np.array([0, 0, 0, 1])
    return

def setPlayerTeam(player, teamID, playerList, gameLineup):
    if player not in playerList:
        #Team event
        playerTeam = teamID
    else:
        playerTeam = gameLineup[gameLineup['Person_id'] == player]['Team_id'].values[0]
    return playerTeam
```

```python
game0 = GameIDs[0]
game0PlayByPlay= sortedPlayByPlay(PlayByPlay[PlayByPlay['Game_id'] == game0])
game0Lineup = GameLineup[GameLineup['Game_id'] == game0]
game0Teams = list(game0Lineup['Team_id'].unique())

#Dictionary of every player (active or inactive for this game)
game0Dict = {player: np.array([0,0,0]) for player in list(game0Lineup['Person_id'].unique())}

#Following are the starting lineups for each time
team0Lineup = list(game0Lineup[(game0Lineup['Period'] ==1) & (game0Lineup['Team_id'] == game0Teams[0])]['Person_id'])
team1Lineup = list(game0Lineup[(game0Lineup['Period'] ==1) & (game0Lineup['Team_id'] == game0Teams[1])]['Person_id'])
ftTeam0Lineup = team0Lineup
ftTeam1Lineup = team1Lineup
```

```python
def gameRatings(gamePlayByPlay, gameLineup, eventCodes):
    lastDescript = ''
    counter = 0
    gameTeams = list(gameLineup['Team_id'].unique())
    
    #Create dictionary with keys of each player and value of array([PointsFor,PointsAgainst, OffPossessions, DefPossessions])
    playerList = list(gameLineup['Person_id'].unique())
    gameDict = {player: np.array([0,0,0,0]) for player in playerList}
    
    #List of event numbers to skip (designed for the case where we 'look ahead' at a future rebound due to playbyplay error)
    skipEventNums = []
    
    for event in gamePlayByPlay.itertuples():
        #print('Counter', counter)
        #print('Event Num', getattr(event, 'Event_Num'))
        #Get the event type
        if getattr(event, 'Event_Num') in skipEventNums:
            continue
        
        eventMsgType = getattr(event, "Event_Msg_Type")
        actionType = getattr(event, 'Action_Type')
        
        #Turn the event type and action type into a description
        #First find corresponding row in eventCodes file
        eventCode = eventCodes[(eventCodes['Event_Msg_Type'] == eventMsgType) & (eventCodes['Action_Type'] == actionType)]

        #Description should be unique based off ID, so just get first element of list created
        eventMsgDescript = list(eventCode['Event_Msg_Type_Description'].unique())[0]
        actionTypeDescript = list(eventCode['Action_Type_Description'].unique())[0]

        player = getattr(event, 'Person1') #Player who shoots, rebounds, etc
        
        teamID = getattr(event, 'Team_id') #teamID doesn't always correspond to player's team
        playerTeam = setPlayerTeam(player, teamID, playerList, gameLineup)
        #if counter == 128:
        #    print(team0Lineup)
        #    print(team1Lineup)
        #    print('player:', player)
        #    print('descript', lastDescript, eventMsgDescript)
        #Make sure none of the following gets double counted -- most likely on and-1s

        #Substitutions at start of period aren't considered substitution events. Must check game linup explicitly. 
        if eventMsgDescript == 'Start Period':
            period = getattr(event, 'Period')
            team0Lineup = list(gameLineup[(gameLineup['Period'] == period) & (gameLineup['Team_id'] == gameTeams[0])]['Person_id'])
            team1Lineup = list(gameLineup[(gameLineup['Period'] == period) & (gameLineup['Team_id'] == gameTeams[1])]['Person_id'])
            #print(team0Lineup)
            #print(team1Lineup)
            ftTeam0Lineup = team0Lineup
            ftTeam1Lineup = team1Lineup
            lastDescript = ''
        elif eventMsgDescript == 'End Period':
            #Add possession to each player
            if playerTeam == gameTeams[0]:
                #team0 on offense
                addPossession(team0Lineup, team1Lineup, gameDict)
            else:
                addPossession(team1Lineup, team0Lineup, gameDict)
            lastDescript = ''
        if lastDescript == 'Missed Shot':
            
            #Noticed some stoppage (probably clock malfunction after missed shot, the next event was a rebound)
            #See game1, eventnum 392, 401, 394
            #if eventMsgDescript == 'Stoppage':
            #    continue
            #if eventMsgDescript == 'Missed Shot':
                #Noticed error where a missed shot recorded then immediately followed by miss/made shot, 
                #then followed by rebound
                #See game6 event_num 374, 383, 375, 385 (index 270:290 should cover)
                #This must imply offensive rebound
            #    continue
            #if eventMsgDescript == 'Made Shot':
                #Noticed two missed shots in a row, so adding this in case second shot made w/ same error in reporting
            #    points = getattr(event, 'Option1')
            #    addPoints(player,points, team0Lineup, team1Lineup, gameDict, possession=1)
            #    lastDescript = ''
            #    continue
                
            lastDescript = ''
            if eventMsgDescript != 'Rebound':
                #Does a missed shot that goes out of bounds count as turnover? Don't think so
                
                #Errors in playbyplay sometimes lead to the wrong order if two things happen at same playclock time
                #Check other events at same time to see if rebound occurs
                
                PCtime = getattr(event, 'PC_Time')
                period = getattr(event, 'Period')
                
                #Check if rebound occurs at same time (period and playclock time, arena time does not work!)
                reboundEvent = gamePlayByPlay[(gamePlayByPlay['Period'] == period) & (gamePlayByPlay['PC_Time'] <= PCtime) & (gamePlayByPlay['Event_Msg_Type'] == 4)]
                if reboundEvent.empty:
                    print('No rebound at same time!')
                    print('Counter', counter)
                    print('Event num', getattr(event, 'Event_Num'))
                    break
                #Get first instance of rebound
                player = reboundEvent['Person1'].values[0]
                playerTeam = setPlayerTeam(player, reboundEvent['Team_id'].values[0], playerList, gameLineup)
                
                #Make sure to skip this rebound event
                skipEventNums.append(reboundEvent['Event_Num'].values[0])
                #print('missed shot')
                #print('Counter', counter)
                #print('Event num', getattr(event, 'Event_Num'))
                #print("Rebound did not occur after missed shot!")
                #break
            #player = getattr(event, 'Person1') #Person who grabbed rebound
            
            if player not in lastTeam: #Defensive rebound (for an offensive rebound do nothing)
                if (player not in team0Lineup) and (player not in team1Lineup) and (player in playerList):
                    print('Player not on court', player)
                    print('team0', team0Lineup)
                    print('team1', team1Lineup)
                    print('Counter', counter)
                    print('Event num', getattr(event, 'Event_Num'))
                if playerTeam == gameTeams[0]:
                    #team0 on offense
                    addPossession(team0Lineup, team1Lineup, gameDict)
                else:
                    addPossession(team1Lineup, team0Lineup, gameDict)
                
            #Reset to player of current event
            player = getattr(event, 'Person1')
            playerTeam = setPlayerTeam(player, teamID, playerList, gameLineup)
        elif lastDescript == 'Missed Last Free Throw':
            #Noticed some stoppage (probably clock malfunction after missed shot, the next event was a rebound)
            #if eventMsgDescript == 'Stoppage':
            #    continue
            lastDescript = ''
            if eventMsgDescript != 'Rebound':
                #Does a missed shot that goes out of bounds count as turnover? Don't think so
                
                #Errors in playbyplay sometimes lead to the wrong order if two things happen at same playclock time
                #Check other events at same time to see if rebound occurs
                
                PCtime = getattr(event, 'PC_Time')
                period = getattr(event, 'Period')
                
                #Check if rebound occurs at same time or after (period and playclock time, arena time does not work!)
                #From the instances where I've noticed an issue, the rebound always occurs after missed shot but not necessarily same time
                reboundEvent = gamePlayByPlay[(gamePlayByPlay['Period'] == period) & (gamePlayByPlay['PC_Time'] <= PCtime) & (gamePlayByPlay['Event_Msg_Type'] == 4)]
                if reboundEvent.empty:
                    print('Counter', counter)
                    print('Event num', getattr(event, 'Event_Num'))
                    print('No rebound at same time!')
                    break
                #Get first instance of rebound
                player = reboundEvent['Person1'].values[0]
                playerTeam = setPlayerTeam(player, reboundEvent['Team_id'].values[0], playerList, gameLineup)
                
                #Make sure to skip this rebound event
                skipEventNums.append(reboundEvent['Event_Num'].values[0])
                
                #print('missed shot')
                #print('Counter', counter)
                #print('Event num', getattr(event, 'Event_Num'))
                #print("Rebound did not occur after missed shot!")
                #break
            if player in lastTeam: #Offensive Rebound
                #Add possession to the players who were subbed out, if any
                subbedPlayers0 = list(set(ftTeam0Lineup) - set(team0Lineup))
                subbedPlayers1 = list(set(ftTeam1Lineup) - set(team1Lineup))
                if (len(subbedPlayers0) > 0) or (len(subbedPlayers1) > 0):
                    if playerTeam == gameTeams[0]:
                        #team0 on offense
                        addPossession(subbedPlayers0, subbedPlayers1, gameDict)
                    else:
                        addPossession(subbedPlayers1, subbedPlayers0, gameDict)
 
            else: #Defensive Rebound
                #Only add possession to those who were on court when foul occured
                if playerTeam == gameTeams[0]:
                    #team0 on offense
                    addPossession(ftTeam0Lineup, ftTeam1Lineup, gameDict)
                else:
                    addPossession(ftTeam1Lineup, ftTeam0Lineup, gameDict)
            #Reset to player of current event
            player = getattr(event, 'Person1')
            playerTeam = setPlayerTeam(player, teamID, playerList, gameLineup)

        if eventMsgDescript == 'Made Shot':
            #player = getattr(event, 'Person1') #Player who makes a shot
            points = getattr(event, 'Option1')
            addPoints(player,points, team0Lineup, team1Lineup, gameDict, possession=1)
        elif eventMsgDescript == 'Turnover':
            #Add possession to each player
            if playerTeam == gameTeams[0]:
                #team0 on offense
                addPossession(team0Lineup, team1Lineup, gameDict)
            else:
                addPossession(team1Lineup, team0Lineup, gameDict)
        elif eventMsgDescript == 'Missed Shot':
            #Check who rebounds in next iteration
            lastDescript = eventMsgDescript
            #player = getattr(event, 'Person1') #Player who misses shot
            if player in team0Lineup:
                lastTeam = team0Lineup
            else:
                lastTeam = team1Lineup

        #Check if Foul (which may or may not lead to free throws), but all free throws come from fouls
        #Important to check if a generic foul and not just shooting fouls, otherwise the free throw lineup will be 'behind'
        #If only looked at shooting fouls - consider situation of substitutions during the game and then a technical foul. 
        #The ft lineup would only be from the previous fts, not the current lineup. 
        elif (eventMsgDescript == 'Foul'):
            ftTeam0Lineup = team0Lineup
            ftTeam1Lineup = team1Lineup
        elif eventMsgDescript == 'Free Throw':
            option1 = getattr(event, 'Option1')
            #player = getattr(event, 'Person1') #Player who takes free throw
            if option1 == 1: #Made Free Throw
                points = 1
                if (actionType == 10) or (actionType == 12) or (actionType == 15):
                    #ActionTypes corresponding to final free throw, so add possession. Also update ft lineups to on court lineup
                    addPoints(player, points, ftTeam0Lineup, ftTeam1Lineup, gameDict, possession=1)
                    ftTeam0Lineup = team0Lineup
                    ftTeam1Lineup = team1Lineup
                else:
                    addPoints(player, points, ftTeam0Lineup, ftTeam1Lineup, gameDict, possession=0)
            else: #Missed FT
                if (actionType == 10) or (actionType == 12) or (actionType == 15):
                    #Missed last free throw for a shooting foul (not technical/flagrant), check who rebounds in next iteration. 
                    lastDescript = 'Missed Last Free Trow'
                    #Offensive rebounds with substitution going to be weird in terms of possessions

                    #Check current lineup, not free throw lineup
                    if player in team0Lineup:
                        lastTeam = team0Lineup
                    else:
                        lastTeam = team1Lineup


            #do harder stuff
        elif eventMsgDescript == 'Substitution':
            leavingPlayer = player
            #if player == '4dd3d6a51dc97c651d3a86eec4362a1f':
            #    print('being subbed', counter)
            enteringPlayer = getattr(event, 'Person2')
            if leavingPlayer in team0Lineup:
                team0Lineup.remove(leavingPlayer)
                team0Lineup.append(enteringPlayer)
            else:
                team1Lineup.remove(leavingPlayer)
                team1Lineup.append(enteringPlayer)

        counter += 1
    return gameDict
```

```python
dataRows = []
i = 0
for game in GameIDs:
    print(i)
    gamePlayByPlay= sortedPlayByPlay(PlayByPlay[PlayByPlay['Game_id'] == game])
    gameLineup = GameLineup[GameLineup['Game_id'] == game]
    gameDict = gameRatings(gamePlayByPlay, gameLineup, eventCodes)
    for player in gameDict.keys():
        dataDict = {'Game_ID': game}
        #Check at least one possession played
        possessions = gameDict[player][2]
        if possessions != 0:
            pointsFor = gameDict[player][0]
            pointsAgainst = gameDict[player][1]
            dataDict.update({'Player_ID': player, 'OffRtg': 100*(pointsFor/possessions), 'DefRtg': 100*(pointsAgainst/possessions), 'Possessions': possessions})
            
            dataRows.append(dataDict)
    i+=1
df = pd.DataFrame(dataRows)
df= df[['Game_ID', 'Player_ID', 'OffRtg', 'DefRtg', 'Possessions']]
df.to_excel("output.xlsx", index=False)
```
